```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('WhatsApp Image 2024-09-24 at 11.02.14_98e6ee12.jpg')
# take your own image path .
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
```

```python
edges = cv2.Canny(blurred_image,100,200)
```

```python
titles = ['Original Image','Grayscale','Blurred','Edages']
images = [image,gray_image,blurred_image,edges]
```

```python
for i in range(4):
    plt.subplot(2,2,i+1)
    if i ==0:
        plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])
        plt.show()
```
